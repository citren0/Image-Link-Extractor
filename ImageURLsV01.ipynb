```python
#data = pd.read_csv('Core_Dataset.csv', dtype=object)
data = pd.read_csv('Core_Dataset_5only.csv', dtype=object)
#data = pd.read_csv('Core_Dataset_100only.csv', dtype=object)
#print(data.loc[3, 'Website'])
print(data)
```

```python
import csv
with open('OUTPUT_AdditionalURLs.csv', mode='w') as file:
    writer = csv.writer(file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    
    writer.writerow(['ContactID', 'OrganizationName', 'URLUsed', 'altImageDescription', 'Found_IMG_URL'])
    # 
    anchortags= []
    # taking 
    for index,row in data.iterrows():
        orgID = data.loc[index, 'ContactID']
        orgName = data.loc[index, 'OrganizationName']
        orgURL = data.loc[index, 'Website']
        # make sure there is access to input URL
        try:
            page = requests.get(orgURL)     #to extract page from website
            html_code = page.content        #to extract html code from page
        except Exception as e:
            print(e)
        # store all image tags
        try:
            soup = BeautifulSoup(html_code, 'html.parser')  #Parse html code / cleans all HTML to make it readable
            anchortags = soup.find_all(['img'])             #find all IMG tags
        except Exception as e:
            print(e)
        # for every image, get the image value for URL path + alternative description
        for link in soup.find_all('img'):
            foundURL = link.get('src')  # get the URL path of the image contained in the src attribute
            try:
                altImageDescription = link.get('alt').strip()    # get the alternative description of the image 
            except Exception as e:
                print("No alt text found.")
            # Initialize keep attribute to true at first   
            keep = True
            # If src link has a value, decoding URL bytestring to unicode string 
            if foundURL:
                foundURL = bytes(foundURL, 'utf-8').decode('utf-8', 'ignore')
            # If src is null, do not keep value 
            if not foundURL:
                keep = False
            # If first character found is a '#', do not keep link
            elif foundURL[0] == '#':
                keep = False
            # If double backslash '\\' is found anywhere in foundURL, do not keep link
            elif "\\" in foundURL:
                keep = False
            # If first character is a front slash, it means it's a partial image URL path 
            elif foundURL[0] == '/':
                try:
                    completeURL = []
                    completeURL.append(orgURL)   # orgURL is base website URL from input CVS file
                    completeURL.append(foundURL) # foundURL is the partial image URL path
                    foundURL = ''.join(completeURL)
                    #foundURL = orgURL + foundURL was giving me errors for adding floats and strings
                except Exception as e:
                    print("None found.")
                    
            if "jpg" in foundURL or "png" in foundURL:
                keep = True
            else:
                keep = False
            if keep:    
                print('orgID: ', orgID)
                print('orgName: ', orgURL)
                print('orgURL: ', orgName)
                print('altImageDescription: ', altImageDescription)
                print('foundURL: ', foundURL)
                print('\n')
                  
                try:
                    #Write results to output CSV
                    writer.writerow([orgID, orgName, orgURL, altImageDescription, foundURL])
                except Exception as e:
                    print(e)
    
```
